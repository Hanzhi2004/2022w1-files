```python
from typing import NamedTuple, List

GovernmentGrant = NamedTuple('GovernmentGrant', [('city', str),      
                                                     ('org', str),   
                                                     ('type', str), 
                                                     ('area', str),    
                                                     ('subarea', str),      
                                                     ('amt', int)])     # in range[0, ...)
# interp. government grant data from BC. includes the city, organization ('org'), grant type ('type'),
#         grant area ('area'), grant subarea ('subarea'), and payment amount ('amt') in CAD

AGRIFAIR = GovernmentGrant("Abbotsford", 
                                "Abbotsford Agrifair", 
                                "Government Grant", 
                                "Arts and Culture", 
                                "Fairs, Festivals, Museums", 
                                80000)
BMX = GovernmentGrant("Abbotsford", 
                           "Abbotsford BMX Society", 
                           "Government Grant", 
                           "Sport", 
                           "Sports for Youth", 
                           29500)
BIA = GovernmentGrant("Abbotsford", 
                           "Fraser Valley Brain Injury Association", 
                           "Government Grant", 
                           "Human and Social Services", 
                           "Disadvantage Distress Poverty", 
                           60000)
AIRSHOW = GovernmentGrant("Abbotsford", 
                               "Abbotsford International Airshow Society", 
                               "Government Grant", 
                               "Arts and Culture", 
                               "Fairs, Festivals, Museums", 
                               60000)
ARTS_COUNCIL = GovernmentGrant("Burnaby",
                                    "Burnaby Arts Council", 
                                    "Government Grant", 
                                    "Arts and Culture", 
                                    "Adult Arts and Culture", 
                                    15100)
BARRACUDAS = GovernmentGrant("Burnaby", 
                                  "Burnaby Barracuda Summer Aquatics Club", 
                                  "Government Grant", 
                                  "Sport", 
                                  "Sports for Youth", 
                                  51900)
AFRICAN_DESCENT_SOCIETY = GovernmentGrant("Vancouver", 
                                               "African Descent Society British Columbia", 
                                               "Government Grant", 
                                               "Arts and Culture", 
                                               "Fairs, Festivals, Museums", 
                                               25000)
ALL_BODIES = GovernmentGrant("Vancouver", 
                                  "All Bodies Dance Society", 
                                  "Government Grant", 
                                  "Arts and Culture", 
                                  "Adult Arts and Culture",   
                                  12000)
YOUTH_MUSIC = GovernmentGrant("Richmond", 
                                   "BC Youth Music Society", 
                                   "Government Grant", 
                                   "Arts and Culture", 
                                   "Adult Arts and Culture",   
                                   31500)
CHINESE_ORCHESTRA = GovernmentGrant("Richmond", 
                                         "Canadian YC Chinese Orchestra Association",
                                         "Government Grant", 
                                         "Arts and Culture", 
                                         "Adult Arts and Culture",   
                                         5000)

# template based on compound
def fn_for_government_grant(gg: GovernmentGrant) -> ...:
    return ...(gg.city,
               gg.org,
               gg.type,
               gg.area,
               gg.subarea,
               gg.amt)


# List[GovernmentGrant]
# interp. a list of government grants

L0 = []
L1 = [AGRIFAIR, 
      BMX, 
      BIA, 
      AIRSHOW, 
      ARTS_COUNCIL, 
      BARRACUDAS, 
      AFRICAN_DESCENT_SOCIETY, 
      ALL_BODIES, 
      YOUTH_MUSIC, 
      CHINESE_ORCHESTRA]

# template based on arbitrary-sized and the reference rule
def fn_for_logg(logg: List[GovernmentGrant]) -> ...:
    # description of the acc
    acc = ... # type: ...
    for gg in logg:
        acc = ...(acc, fn_for_government_grant(gg))
    return ...(acc)
```

## Problem 1

Design a function that takes a city, a grant amount, and a list of government grants and returns **all the government grants** awarded in that city of at most the given grant amount (at most = smaller or equal).

In particular, the **function MUST operate in two steps**. First, it will find the government grants in the given city, then it will find which of those government grants received at most the maximum grant amount.

Remember to follow the HtDF recipe and the helper rules.

```python
@typecheck
def filter_on_city_and_grant_amount(city: str, amt: int, logg: List[GovernmentGrant]) -> List[GovernmentGrant]:
    """
    return a list of grants from logg that are from the given city and of at most amt
    """
    #return [] #stub
    # template based on function composition
    
    # You aren't required to include a plan in comments, 
    # and your solution can be all in one line. But..
    # we like the way this looks and reads!
    
    
    # Plan:
    # 1) find the government grants in logg from given city, call it city_grants
    # 2) find the government grants from city_grants with grant size at most amt, call it small_city_grants
    # 3) return small_city_grants
    
    city_grants = from_city_only(city, logg)
    small_city_grants = smaller_than_only(amt, city_grants)
    return small_city_grants
    
    # The alternate one-line version would be:
    #return smaller_than_only(amt, from_city_only(city, logg))
    
    # Can you see how that looks like the three lines above,
    # but with each variable name "substitude" for the expression
    # from which we got its value?

@typecheck
def smaller_than_only(amt: int, logg: List[GovernmentGrant]) -> List[GovernmentGrant]:
    """
    return a list of grants from logg that are of at most amt
    """
    # return [] #stub
    # template from List[GovernmentGrant] with additional parameter
    
    # acc stores the result so far
    acc = [] # type: List[GovernmentGrant]
    
    for gg in logg:
        if grant_smaller_or_equal_than(amt, gg):
            acc.append(gg)
    
    return acc

@typecheck
def grant_smaller_or_equal_than(amt: int, gg: GovernmentGrant) -> bool:
    """
    return True if gg has an amount smaller or equal than amt and False otherwise
    """
    #return False #stub
    # template from GovernmentGrant with additional parameter
    return gg.amt <= amt

@typecheck
def from_city_only(city: str, logg: List[GovernmentGrant]) -> List[GovernmentGrant]:
    """
    return a list of grants from logg that are from a given city
    """
    # return [] #stub
    # template from List[GovernmentGrant] with additional parameter
    
    # acc stores the result so far
    acc = [] # type: List[GovernmentGrant]
    
    for gg in logg:
        # It is perhaps a judgment call whether we need
        # a helper to check if gg is from area, but in
        # CPSC 103, we ask you to create this helper!
        if from_city(city, gg):
            acc.append(gg)
    
    return acc

@typecheck
def from_city(city: str, gg: GovernmentGrant) -> bool:
    """
    return True if gg is from the given city, False otherwise
    """
    #return False #stub
    # template from GovernmentGrant with additional parameter
    return gg.city == city

start_testing()

expect(filter_on_city_and_grant_amount("Vancouver", 20000, []), []) 
expect(filter_on_city_and_grant_amount("Vancouver", 20000, L1), [ALL_BODIES])
expect(filter_on_city_and_grant_amount("Abbotsford", 79999, L1), [BMX, BIA, AIRSHOW])
expect(filter_on_city_and_grant_amount("Abbotsford", 80000, L1), [AGRIFAIR, BMX, BIA, AIRSHOW])
expect(filter_on_city_and_grant_amount("Richmond", 30000, L1), [CHINESE_ORCHESTRA])
expect(filter_on_city_and_grant_amount("Richmond", 31500, L1), [YOUTH_MUSIC, CHINESE_ORCHESTRA])
expect(filter_on_city_and_grant_amount("Richmond", 3000, L1), [])


expect(smaller_than_only(1, []), [])
expect(smaller_than_only(1, L1), [])
expect(smaller_than_only(25000, L1), [ARTS_COUNCIL, AFRICAN_DESCENT_SOCIETY, ALL_BODIES, CHINESE_ORCHESTRA])
expect(smaller_than_only(100000, L1), L1)

expect(grant_smaller_or_equal_than(3000, CHINESE_ORCHESTRA), False)
expect(grant_smaller_or_equal_than(5000, CHINESE_ORCHESTRA), True)
expect(grant_smaller_or_equal_than(4999, CHINESE_ORCHESTRA), False)
expect(grant_smaller_or_equal_than(29499, BMX), False)
expect(grant_smaller_or_equal_than(29500, BMX), True)
expect(grant_smaller_or_equal_than(29501, BMX), True)

expect(from_city_only("Abbotsford", []), []) 
expect(from_city_only("Abbotsford", L1), [AGRIFAIR, BMX, BIA, AIRSHOW])
expect(from_city_only("Vancouver", L1), [AFRICAN_DESCENT_SOCIETY, ALL_BODIES])
expect(from_city_only("Toronto", L1), [])

expect(from_city("Richmond", CHINESE_ORCHESTRA), True)
expect(from_city("Vancouver", CHINESE_ORCHESTRA), False)
expect(from_city("Abbotsford", BMX), True)
expect(from_city("Vancouver", BMX), False)

summary()
```
